```python
import numpy as np
# conditionals
# if, elif, else
# looping
```

```python
x = 5

if(x == 5):
    print("I'm 5!")
```

```python
x = 5

if(x == 5):
    print("I'm 5!")
else:
    print("I'm not 5!")
    
```

```python
x = 4

if(x == 5):
    print("I'm 5!")
elif(x == 4):
    print("I'm 4!")
else:
    print("I'm not 5!")
    
```

```python
# once it finds first true it exits out

x = 4

if(x > 3):
    print("I'm greater than 3")
elif(x == 4):
    print("I'm 4!")

# this is called short circuiting 
```

```python
arr = np.arange(0, 10)

# looping over a container (list or array)
li = ["h", "e", "l", "l", "o", "w", "o", "r", "l", "d", "!"]
for i in li:
    print(i)

x = np.zeros(len(arr))
print(x)
for i in arr:
    x[i] += 5
    x[i] += arr[i]

print(x)
# looping over a container and modifying it
# for i in arr

# looping until condition is met
# if you're not careful this can run "forver"
# count = 0
# while(count < 5):
#     print(count)
#     count += 1
    
```
