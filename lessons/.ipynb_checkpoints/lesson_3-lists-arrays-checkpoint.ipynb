# Lists

- containers that hold variables
- the variables can be a mix of things


```python
# lists
list1 = [0, 1.0, "hello", 10, "world"]
list2 = [0, 0, 0, 0]
list1
```

```python
list1[0:5]
```

accessing an element of list

Python indexes at 0, so first element of a list is list1[0]

accessing a portion of list (also called slicing)

```python
list1[0:2]
```

can do math with lists and treat them like variables

```python
list1
```

```python
z = list1[1] + 3
z**2
```

can also append to a list

```python
list1
```

```python
list1.append(0)
```

```python
list1
```

# Arrays

- Slightly different from lists
- are imported from `numpy`
- can only have numbers in them
- can not "expand" i.e. `append` doesn't work
- can handle "math" operations better/faster
- are faster in general
- have inherit attributes (like size) (Table 4.1)

```python
import numpy as np
```

```python
arr = np.array([1, 2, 3])
arr
```

There are various ways to make arrays
- np.linspace or logspace (start, stop, N)
- np.zeros
- np.arange (start, stop, step)
- np.ones

Pg 63 of book has more details

```python
np.linspace(0,10,5)
```

Can do more math operations on the entire array

slicing is probably the most powerful thing but also the most confusing when you are first using it


```python
arr
```

```python
arr = np.linspace(0, 5, 6)
arr[0:3]
```

```python
np.ones(6)
```

```python
arr
```

```python
arr = np.ones((6,6))
# arr[0:3,:] *= 2
arr[0:3,:] + 3
```
